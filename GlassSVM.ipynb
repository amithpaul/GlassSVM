```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

glassdata=pd.read_csv("/content/drive/MyDrive/glass.csv")
```

```python
glassdata.head() #shows first 5 records
```

## Pre-Processing

```python
glassdata.describe() #gives aggregate function and other details results for each feature
```

From the description we can say that Ba and Fe can be exclued as most of the values (75% row of result from descibe functiion) for these fields are zero and are thus irrelevant as a costituent of glass.

```python
print((glassdata['Ba']==0).sum()/len(glassdata))
print((glassdata['Fe']==0).sum()/len(glassdata))
```

```python
# Deletes the irrelevant features
glassdata.drop('Ba',axis=1,inplace=True)
glassdata.drop('Fe',axis=1,inplace=True)
```

```python
# Print altered dataset
print(glassdata)
```

```python
from sklearn.model_selection import train_test_split

X = glassdata.iloc[:,:-1] # Takes all columns except the last column
Y = glassdata.iloc[:,-1] # Takes only the last column (Type)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# print(X_train.shape)
# print(X_test.shape)
# print(Y_train.shape)
# print(Y_test.shape)
```

## Training

```python
from sklearn.svm import SVC

model = SVC(kernel='linear',C=0.8)
model.fit(X_train, Y_train)

Y_predict=model.predict(X_test)
```

### Metrics

#### Accuracy score

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(Y_test,Y_predict))
```

#### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(Y_test,Y_predict))
```

#### Classification Report

```python
from sklearn.metrics import classification_report

print(classification_report(Y_test,Y_predict))
```
